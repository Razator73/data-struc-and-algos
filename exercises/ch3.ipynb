### R-3.1
Graph the functions $8n$, $4n\,log(n)$, $2n^2$, $n^3$, and $2^n$ using a logarithmic scale
for the $x$- and $y$-axes; that is, if the function value $f(n)$ is $y$, plot this as a
point with $x$-coordinate at $log\,n$ and $y$-coordinate at $log\,y$

```python
pts = [2**(k/2) for k in range(10)]
x = [n for n in pts]
y_1 = [8*n for n in pts]
y_2 = [4*n*math.log(n) for n in pts]
y_3 = [2*n**2 for n in pts]
y_4 = [n**3 for n in pts]
y_5 = [2**n for n in pts]

plt.yscale('log')
plt.xscale('log')
plt.plot(x, y_1, label=r'$f(x)=8n$')
plt.plot(x[1:], y_2[1:], label=r'$f(x)=4n\,log(n)$')
plt.plot(x, y_3, label=r'$f(x)=n^2$')
plt.plot(x, y_4, label=r'$f(x)=n^3$')
plt.plot(x, y_5, label=r'$f(x)=2^n$')
plt.legend()
plt.show()
```

### R-3.2
The number of operations executed by algorithms A and B is $8n\,log\,n$ and
$2n^2$, respectively. Determine $n_0$ such that A is better than B for $n \ge n_0$.

#### Answer
Using the code below we can see that for $n\ge8$ that A will be better than B

```python
x = [n for n in range(1, 20)]
y_1 = [8*n*math.log(n) for n in x]
y_2 = [2*n**2 for n in x]

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.show()

for i in range(len(x)):
    print(i, y_1[i] < y_2[i])
```

### R-3.5
Explain why the plot of the function nc is a straight line with slope c on a
log-log scale.

#### Answer
For $f(n) = n^c$ then applying a log scale to both axes will be the same as saying 
$log(f(n)) = log(n^c)$ then applying log properties would give $log(f(n)) = c\,log(n)$ 
thus making the function look linear with a slope of $c$

### R-3.6
What is the sum of all the even numbers from $0$ to $2n$, for any positive
integer $n$?

#### Answer
Note $0$ is dropped as it doesn't add anything to the summation when included. Then 
based on Proposition 3.3 we can give the following summation a function.

\$
\displaystyle\sum_1^n2i = 2\displaystyle\sum_1^ni = 
2\frac{n(n+1)}{2} = n(n + 1) = n^2 + n
\$

See below for some randomly tested integers


```python
n_vals = [random.randint(1, 800) for i in range(10)]
n_sums = [str(sum([x for x in range(2, 2*n + 1, 2)])) for n in n_vals]
n_funcs = [str(n**2 + n) for n in n_vals]
n_vals = [str(n) for n in n_vals]
max_val_len = max([len(n) for n in n_vals])
max_sum_len = max([len(n) for n in n_sums])
max_func_len = max([len(n) for n in n_funcs])
for i in range(len(n_vals)):
    val_str = f'for n = {" "*(max_val_len - len(n_vals[i]))}{n_vals[i]}'
    func_str = f'n^2 + n = {" "*(max_func_len - len(n_funcs[i]))}{n_funcs[i]}'
    sum_str = f'sum = {" "*(max_sum_len - len(n_sums[i]))}{n_sums[i]}'
    print(f'{val_str}, {func_str} and {sum_str}' )
    assert n_funcs[i] == n_sums[i]
```
